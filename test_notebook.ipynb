```python
from pathlib import Path
from typing import TypedDict, List
from grpo_rag.utils.constants import MEMORY_BUFFER_PATH
```

```python
class MemoryEntry(TypedDict):
    query: str
    answers: List[str]
    advantage_scores: List[float]
```

```python
def append_memory_entry(query, answers, advantage_scores):
    memory_entry = MemoryEntry(
        query=query,
        answers=answers,
        advantage_scores=advantage_scores
    )
    append_jsonl(MEMORY_BUFFER_PATH, memory_entry)
```

```python
append_memory_entry(query='tell me a joke', answers=test_rollouts, advantage_scores=advantage_scores)
```
